### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
def Range_Column(df,column):
     if (type(df.loc[0,column]) is not (np.int64)) & (type(df.loc[0,column]) is not (np.float64))  :
        pass
     else:
        x=(df[column].max()-df[column].min())
        return x

for key,value in Iris.iteritems():
    print('Column: ',key,'\t\t\tRange: ',Range_Column(Iris,key))
    
```
